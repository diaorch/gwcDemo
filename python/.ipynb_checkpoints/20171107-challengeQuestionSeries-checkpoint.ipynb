Plotting centroids

```python
centroid_coordinates.values()
```

```python
plotData = {'x':[], 'y':[], 'label':[]}
for state, coordinates in centroid_coordinates.items():
    plotData["y"].append(coordinates[0])
    plotData["x"].append(coordinates[1])
    plotData["label"].append(state)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# display scatter plot data
plt.figure(figsize=(16,9))
plt.title('State centroids', fontsize=20)
plt.xlabel('W', fontsize=15)
plt.ylabel('N', fontsize=15)
plt.scatter(plotData["x"], plotData["y"], marker = 'o')
plt.axis([max(plotData["x"]), min(plotData["x"]), min(plotData["y"]), max(plotData["y"])])

# add label
for label, x, y in zip(plotData["label"], plotData["x"], plotData["y"]):
    plt.annotate(label, xy = (x, y))
```
